```python
import csv, urllib
```

```python
gemeenten = []

with open('gemeenten.csv') as r:
    reader = csv.reader(r, delimiter=';')
    for l in reader:
        gemeenten.append(l[0])

gemeenten
```

```python
for gemeente in gemeenten:
    with open('data/%s.xml' % gemeente, 'w') as w:
        url = 'http://novumuitslagencentrum.nl/gemeenten/%s.xml' % urllib.quote(gemeente)
        print 'loading and writing %s' % url
        doc = urllib.urlopen(url).read()
        w.write(doc)
```

```python
from bs4 import BeautifulSoup
from decimal import Decimal
import json
from json import encoder
encoder.FLOAT_REPR = lambda o: format(o, '.6f')

partijen = ["VVD", "PvdA", "PVV", "SP", "CDA", "D66", "CU", "GrLinks", "SGP", "PvdD", "50+"]

uitslagen = []

for gemeente in gemeenten:
    with open('data/%s.xml' % gemeente) as g:
        uitslag = {}

        soup = BeautifulSoup(g.read())
        uitslag['gem'] = soup.uitslag.title.naam.string
        uitslag['prov'] = soup.uitslag.provincie.string
        uitslag['inw'] = int(soup.uitslag.aantalinwoners.string)

        for groep in soup.uitslag.find_all('groep', attrs={'type':'Totaal'}):
            if groep.naam.string == 'Kiesgerechtigden':
                uitslag['kg'] = int(groep.find_all('verkiezing', attrs={'code':'TK12'})[0].stemmen.string)
            else:
                uitslag[groep.kortenaam.string.lower()[:3].encode('ascii')] = float(groep.find_all('verkiezing', attrs={'code':'TK12'})[0].perc.string.replace(',', '.'))
        
        stemmen = {}
        
        for groep in soup.uitslag.find_all('groep', attrs={'type':'Partij'}):
            resultaat = groep.find_all('verkiezing', attrs={'code':'TK12'})[0].perc.string.replace(',', '.')
            stemmen[groep.kortenaam.string.encode('ascii')] = float(resultaat) if resultaat != '-' else 0.0
            
        stemmen_sorted = []
        for p in partijen:
            stemmen_sorted.append(stemmen[p])
            
        uitslag['stm'] = stemmen_sorted
        
        uitslagen.append(uitslag)
        
for expected_u in uitslagen:
    #if i > 50:
    #    continue
    chi_data = []
    for observed_u in uitslagen:
        chi = 0
        for k in range(0, len(partijen)):
            e = float(expected_u['stm'][k])
            o = float(observed_u['stm'][k])
            #chi += 0.0 if e == 0 else math.pow((o - e), 2) / e
            chi += math.pow((o - e), 2)
        chi_data.append(int(math.sqrt(chi)))
    #i += 1
    expected_u['chi'] = chi_data
    
provs = {'NH': 'Noord-Holland', 'ZL': 'Zeeland', 'FR': 'Friesland', 'ZH': 'Zuid-Holland', 'GR': 'Groningen', 'FL': 'Flevoland', 'NB': 'Noord-Brabant', 'LI': 'Limburg', 'GE': 'Gelderland', 'UT': 'Utrecht', 'OV': 'Overijssel', 'DR': 'Drenthe'}

geo = {}

with open('geo.tsv') as r:
    reader = csv.reader(r, delimiter='\t')
    reader.read()
    for r in reader:
        geo[r[2]] = [float(r[0]), float(r[1])]
    
for u in uitslagen:
    u['geo'] = geo['%s, %s, Nederland' % (u['gem'].replace('\'', ''), provs[u['prov']])]

with open('uitslagen.json', 'w') as w:
    w.write(json.dumps(uitslagen, sort_keys=True))
```
